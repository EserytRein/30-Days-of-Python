```python
# -------- FOR LOOP --------
fruits = ['apple', 'banana', 'mango']
for fruit in fruits:
    print(fruit)

for i in range(3):
    print("Index:", i)

for i in range(1, 6):
    print(i)  # 1 to 5

# Loop with index and value
for i, fruit in enumerate(fruits):
    print(i, fruit)

# Loop through dictionary
person = {'name': 'John', 'age': 30}
for key in person:
    print(key, person[key])

for k, v in person.items():
    print(k, v)

# -------- WHILE LOOP --------
count = 0
while count < 3:
    print("Count is", count)
    count += 1

# -------- BREAK & CONTINUE --------
# break stops the loop
for i in range(5):
    if i == 3:
        break
    print("Break at:", i)  # 0 1 2

# continue skips current iteration
for i in range(5):
    if i == 3:
        continue
    print("Continue skips 3:", i)  # 0 1 2 4

# -------- ELSE with LOOPS --------
for i in range(3):
    print(i)
else:
    print("For loop done!")

i = 0
while i < 3:
    print(i)
    i += 1
else:
    print("While loop done!")

# -------- NESTED LOOPS --------
for i in range(2):
    for j in range(2):
        print(f"i={i}, j={j}")

# -------- COMPREHENSIONS --------
squares = [x**2 for x in range(5)]
print("Squares:", squares)

# -------- ZIP: Parallel Iteration --------
names = ['Alice', 'Bob']
scores = [85, 90]
for name, score in zip(names, scores):
    print(name, score)
```

```python
# A higher-order function takes another function as argument or returns a function

# 1. Passing function as argument
def square(x):
    return x * x

def apply(func, value):
    return func(value)

print("Apply square:", apply(square, 5))  # 25

# 2. Returning a function (closure)
def make_multiplier(n):
    def multiplier(x):
        return x * n
    return multiplier

times2 = make_multiplier(2)
times5 = make_multiplier(5)
print("Times 2:", times2(10))   # 20
print("Times 5:", times5(3))    # 15

# 3. Using built-in higher-order functions
numbers = [1, 2, 3, 4, 5]

# map: apply function to each item
squares = list(map(lambda x: x ** 2, numbers))
print("Squares:", squares)

# filter: keep items that match condition
evens = list(filter(lambda x: x % 2 == 0, numbers))
print("Evens:", evens)

# reduce: combine items into one value
from functools import reduce
total = reduce(lambda x, y: x + y, numbers)
print("Total sum:", total)

# 4. Functions as first-class citizens
def greet():
    return "Hello!"

say_hello = greet
print("Greet function:", say_hello())

# 5. HOF with custom logic
def do_twice(func):
    def wrapper(x):
        return func(func(x))
    return wrapper

double = lambda x: x * 2
double_twice = do_twice(double)
print("Double twice 5:", double_twice(5))  # 20
```

```python
import re

# MATCH: checks only at the start of the string
txt = 'I love to teach python and javaScript'
match = re.match('I love to teach', txt, re.I)
if match:
    start, end = match.span()
    print("Match:", txt[start:end])  # I love to teach

match = re.match('I like to teach', txt, re.I)
print("Match result:", match)  # None

# SEARCH: looks through the entire string
txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

search = re.search('first', txt, re.I)
if search:
    start, end = search.span()
    print("Search match:", txt[start:end])  # first

# FINDALL: returns all matches
matches = re.findall('language', txt, re.I)
print("Occurrences of 'language':", matches)

matches = re.findall('python', txt, re.I)
print("Occurrences of 'python':", matches)

matches = re.findall('Python|python', txt)
print("With | operator:", matches)

matches = re.findall('[Pp]ython', txt)
print("With [Pp]:", matches)

# SUB: replacing substrings
txt = '''Python is the most beautiful language that a human being has ever created.
I recommend python for a first programming language'''

replaced = re.sub('Python|python', 'JavaScript', txt, re.I)
print("Replaced:", replaced)

replaced = re.sub('%', '', '''%I a%m te%%a%%che%r% a%n%d %% I l%o%ve te%ach%ing. 
T%he%re i%s n%o%th%ing as r%ewarding a%s e%duc%at%i%ng a%n%d e%m%p%ow%er%ing p%e%o%ple.
I fo%und te%a%ching m%ore i%n%t%er%%es%ting t%h%an any other %jobs. 
D%o%es thi%s m%ot%iv%a%te %y%o%u to b%e a t%e%a%cher?''')
print("Cleaned text:", replaced)

# SPLIT: split based on pattern
txt = '''I am teacher and  I love teaching.
There is nothing as rewarding as educating and empowering people.
I found teaching more interesting than any other jobs.
Does this motivate you to be a teacher?'''
print("Split lines:", re.split('\n', txt))

# Regex pattern declaration
regex_pattern = r'apple'
txt = 'Apple and banana are fruits. An apple a day...'
print(re.findall(regex_pattern, txt, re.I))
print(re.findall(r'[Aa]pple', txt))

# Character classes
txt = 'This example was made on December 6, 2019 and revised on July 8, 2021'
print("All digits:", re.findall(r'\d', txt))
print("All numbers:", re.findall(r'\d+', txt))

# Period .
txt = 'Apple and banana are fruits'
print(re.findall(r'[a].', txt))      # 'an', 'a '
print(re.findall(r'[a].+', txt))     # 'and banana are fruits'
print(re.findall(r'[a].*', txt))     # 'and banana are fruits'

# Optional ? - e.g., optional dash
txt = '''I am not sure if there is a convention how to write the word e-mail.
Some people write it as email others may write it as Email or E-mail.'''
print(re.findall(r'[Ee]-?mail', txt))  # all email variants

# Quantifiers {}
txt = 'This example was made on December 6, 2019 and revised on July 8, 2021'
print(re.findall(r'\d{4}', txt))       # 4-digit numbers
print(re.findall(r'\d{1,4}', txt))     # 1 to 4-digit numbers

# Start of line ^
print(re.findall(r'^This', txt))       # string starts with "This"

# Negation [^...]
print(re.findall(r'[^A-Za-z ]+', txt))  # non-alphabetic characters

# Grouping & alternation
regex_pattern = r'[Aa]pple|[Bb]anana'
txt = 'Apple and banana are fruits. An apple a day...'
print(re.findall(regex_pattern, txt))  # all Apple/banana variants
```

```python
class Person:
    def __init__(self, firstname='Tyrese', lastname='Rein', age=250, country='Finland', city='Helsinki'):
        self.firstname, self.lastname, self.age, self.country, self.city = firstname, lastname, age, country, city
        self.skills = []
    def person_info(self): return f'{self.firstname} {self.lastname} is {self.age} years old. He lives in {self.city}, {self.country}.'
    def add_skill(self, skill): self.skills.append(skill)

class Student(Person):
    def __init__(self, firstname='Tyrese', lastname='Rein', age=250, country='Finland', city='Helsinki', gender='male'):
        self.gender = gender
        super().__init__(firstname, lastname, age, country, city)
    def person_info(self):
        g = 'He' if self.gender == 'male' else 'She'
        return f'{self.firstname} {self.lastname} is {self.age} years old. {g} lives in {self.city}, {self.country}.'

class Statistics:
    def __init__(self, data): self.data = sorted(data)
    def count(self): return len(self.data)
    def sum(self): return sum(self.data)
    def min(self): return min(self.data)
    def max(self): return max(self.data)
    def range(self): return self.max() - self.min()
    def mean(self): return self.sum() / self.count()
    def median(self):
        n = self.count(); mid = n // 2
        return self.data[mid] if n % 2 else (self.data[mid-1] + self.data[mid]) / 2
    def mode(self):
        from collections import Counter
        m = Counter(self.data).most_common(1)[0]
        return m[0], m[1]
    def var(self):
        m = self.mean()
        return sum((x - m) ** 2 for x in self.data) / self.count()
    def std(self): return round(self.var() ** 0.5, 1)
    def freq_dist(self):
        from collections import Counter
        return [(round(v * 100 / self.count(), 1), k) for k, v in Counter(self.data).items()]
    def describe(self):
        return f"""Count: {self.count()}
Sum: {self.sum()}
Min: {self.min()}
Max: {self.max()}
Range: {self.range()}
Mean: {self.mean():.0f}
Median: {self.median()}
Mode: {self.mode()}
Variance: {self.var():.1f}
Standard Deviation: {self.std()}
Frequency Distribution: {self.freq_dist()}"""

# Demo
p1 = Person(); p1.add_skill('HTML'); p1.add_skill('CSS')
print(p1.person_info(), p1.skills)
s1 = Student('Lidiya', 'Teklemariam', 28, 'Finland', 'Espoo', 'female'); s1.add_skill('Marketing')
print(s1.person_info(), s1.skills)

ages = [31,26,34,37,27,26,32,32,26,27,27,24,32,33,27,25,26,38,37,31,34,24,33,29,26]
data = Statistics(ages)
print(data.describe())
```

```python
import json,requests
from bs4 import BeautifulSoup

def clean(t): return ' '.join(t.strip().split())
def table_to_dicts(tbl):
    rows=[[clean(c.get_text(" ",strip=True)) for c in r.find_all(['th','td'])] for r in tbl.find_all('tr')]
    rows=[r for r in rows if any(r)]
    if not rows: return []
    hdr=rows[0]; data=rows[1:] if any(hdr) else (["col"+str(i+1) for i in range(len(rows[0]))],rows)[1]
    if rows and rows[0]==hdr and hdr and hdr[0].lower().startswith('president') and len(hdr)<len(rows[1]): hdr=rows[1]; data=rows[2:]
    return [dict(zip(hdr+[f"col{i+1}" for i in range(len(r)-len(hdr))],r)) for r in data]

def save_json(obj,name):
    with open(name,'w',encoding='utf-8') as f: json.dump(obj,f,ensure_ascii=False,indent=2)
    print(f"Saved {name}")

# 1) BU facts
u1='http://www.bu.edu/president/boston-university-facts-stats/'
s1=BeautifulSoup(requests.get(u1,timeout=30).content,'html.parser')
t1=s1.find_all('table')
bu={"url":u1,"tables":[{"index":i,"caption":clean(t.find('caption').get_text()) if t.find('caption') else None,"rows":table_to_dicts(t)} for i,t in enumerate(t1)]}
save_json(bu,'bu_facts.json')

# 2) UCI datasets table
u2='https://archive.ics.uci.edu/ml/datasets.php'
s2=BeautifulSoup(requests.get(u2,timeout=30).content,'html.parser')
t2=s2.find_all('table',attrs={'cellpadding':'3'})
uci={"url":u2,"table":table_to_dicts(t2[0]) if t2 else []}
save_json(uci,'uci_datasets.json')

# 3) US Presidents table (Wikipedia)
u3='https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States'
s3=BeautifulSoup(requests.get(u3,timeout=30).content,'html.parser')
cand=[t for t in s3.select('table.wikitable')]
pres=None
for t in cand:
    hdr=[clean(th.get_text()) for th in t.find_all('th')]
    if any('President' in h for h in hdr) and any('Term' in h or 'Party' in h for h in hdr): pres=t; break
pres=pres or (cand[0] if cand else None)
wp={"url":u3,"table":table_to_dicts(pres) if pres else []}
save_json(wp,'us_presidents.json')
```

```python
# --- Flask app (single file) ---
from flask import Flask, render_template, request, redirect, url_for
from jinja2 import DictLoader
from collections import Counter
import re, threading, webbrowser

app = Flask(__name__)
app.config['SEND_FILE_MAX_AGE_DEFAULT'] = 0  # avoid CSS caching during dev
app.jinja_loader = DictLoader(TEMPLATES)

@app.route('/')
def home():
    techs = ['HTML', 'CSS', 'Flask', 'Python']
    name = '30 Days Of Python Programming'
    return render_template('home.html', techs=techs, name=name, title='Home')

@app.route('/about')
def about():
    name = '30 Days Of Python Programming'
    return render_template('about.html', name=name, title='About Us')

def analyze_text(text):
    words = re.findall(r"[A-Za-z0-9']+", text.lower())
    counts = Counter(words)
    return {
        "characters": len(text),
        "alnum": sum(ch.isalnum() for ch in text),
        "words": len(words),
        "unique_words": len(counts),
        "top_words": counts.most_common(12),
    }

@app.route('/post', methods=['GET', 'POST'])
def post():
    name = 'Text Analyzer'
    if request.method == 'GET':
        return render_template('post.html', name=name, title=name)
    # POST
    content = request.form.get('content', '')
    stats = analyze_text(content)
    return render_template('result.html', stats=stats, title='Result')

def _open_browser():
    try:
        webbrowser.open("http://127.0.0.1:5000")
    except Exception:
        pass

# Launch the server; this will block the cell. Stop it by interrupting the kernel.
if __name__ == "__main__":
    # Open browser shortly after the server starts
    threading.Timer(1.0, _open_browser).start()
    app.run(host="127.0.0.1", port=5000, debug=False, use_reloader=False)
```

Deleting a document using delete

```python
# let's import the flask
from flask import Flask, Response, request              # >>> added request
from bson.objectid import ObjectId
from bson.json_util import dumps
from datetime import datetime
import os                                               # >>> you use os.environ
import pymongo
# (removed duplicate unused `import json`)              # >>> cleanup

app = Flask(__name__)

MONGODB_URI='mongodb+srv://Tyresekimaru:1234@30daysofpython.9lrojun.mongodb.net/?retryWrites=true&w=majority&appName=30DaysOfPython'
client = pymongo.MongoClient(MONGODB_URI)
db = client['thirty_days_of_python']  # accessing the database

@app.route('/api/v1.0/students', methods=['GET'])
def students():
    cursor = db.students.find()                         # >>> actually query all docs
    return Response(dumps(cursor),                      # >>> return the cursor, not undefined `student`
                    mimetype='application/json')

@app.route('/api/v1.0/students/<id>', methods=['GET'])
def single_student(id):
    # >>> use find_one for a single doc + accept string _id too
    query = {'_id': ObjectId(id)} if ObjectId.is_valid(id) else {'_id': id}
    doc = db.students.find_one(query)
    if not doc:
        return Response(dumps({"error": "not found"}), status=404, mimetype='application/json')
    return Response(dumps(doc), mimetype='application/json')

@app.route('/api/v1.0/students', methods=['POST'])
def create_student():
    name = request.form['name']
    country = request.form['country']
    city = request.form['city']
    skills = request.form['skills'].split(', ')
    bio = request.form['bio']
    birthyear = request.form['birthyear']
    created_at = datetime.now()
    student = {
        'name': name,
        'country': country,
        'city': city,
        'birthyear': birthyear,
        'skills': skills,
        'bio': bio,
        'created_at': created_at
    }
    res = db.students.insert_one(student)               # >>> keep inserted id
    student['_id'] = res.inserted_id                    # >>> include id in response
    return Response(dumps(student), status=201,         # >>> return a response
                    mimetype='application/json')

@app.route('/api/v1.0/students/<id>', methods=['PUT'])
def update_student(id):
    # >>> single PUT route (you had two duplicates); use $set so _id isn't replaced
    query = {'_id': ObjectId(id)} if ObjectId.is_valid(id) else {'_id': id}
    updates = {
        'name': request.form['name'],
        'country': request.form['country'],
        'city': request.form['city'],
        'skills': request.form['skills'].split(', '),
        'bio': request.form['bio'],
        'birthyear': request.form['birthyear'],
        'created_at': datetime.now()
    }
    res = db.students.update_one(query, {'$set': updates})   # >>> use $set
    if res.matched_count == 0:
        return Response(dumps({"error": "not found"}), status=404, mimetype='application/json')
    doc = db.students.find_one(query)                        # >>> return updated doc
    return Response(dumps(doc), mimetype='application/json')

@app.route('/api/v1.0/students/<id>', methods=['DELETE'])
def delete_student(id):
    query = {'_id': ObjectId(id)} if ObjectId.is_valid(id) else {'_id': id}  # >>> string or ObjectId
    res = db.students.delete_one(query)
    if res.deleted_count == 0:
        return Response(dumps({"error": "not found"}), status=404, mimetype='application/json')
    return Response(dumps({"deleted": id}), mimetype='application/json')     # >>> return something

if __name__ == '__main__':
    # for deployment
    # to make it work for both production and development
    port = int(os.environ.get("PORT", 5000))
    app.run(debug=True, host='0.0.0.0', port=port, use_reloader=False)
```
